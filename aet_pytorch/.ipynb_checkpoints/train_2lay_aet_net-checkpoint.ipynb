```python
# Dynamics

# forward dynamics/discretization, euler integration
def euler_dyn(model,input_,params_,t_,alpha_params,DEVICE,inp_on):

    # inputs:
        # input_: input image
        # params_: hyperparameters
        # t_: time vector for integration
        # alpha_params: alpha frequency & amplitude

    # discretization & dynamics parameters
    tau_h,tau_R,R,T,h_start,R_start = params_


    # alpha frequency & amplitude
    _af,_aa = alpha_params

    with torch.no_grad():

        # initialize empty matrices
        dt = np.diff(t_)[0]
        dh1dt = (torch.zeros((model.dims[1],len(t_)+1))*h_start).to(DEVICE)
        dR1dt = (torch.zeros((model.dims[1],len(t_)+1))*R_start).to(DEVICE)
        dh2dt = (torch.zeros((model.dims[2],len(t_)+1))*h_start).to(DEVICE)
        dR2dt = (torch.zeros((model.dims[2],len(t_)+1))*R_start).to(DEVICE)
        dOdt = (torch.zeros((model.dims[-1],len(t_)+1))).to(DEVICE)
        dZdt = (torch.zeros((model.dims[1],len(t_)+1))).to(DEVICE)

        # alpha inhibition
        alpha_inh1 = _aa[0]*np.sin(2*np.pi*_af*t_)+_aa[0]
        alpha_inh2 = _aa[1]*np.sin(2*np.pi*_af*t_-2*np.pi/10)+_aa[1]

        # preactivation (dot product of input and first weight matrix)
        Z,H1,_,_ = model.forw_conv(input_)
        
        # create boxcar function to try different input onsets
        boxcar = np.zeros_like(t_)
        boxcar[inp_on:] = 1

        # adjust initial adaptation term (threshold)
        dR1dt[:,0] = R_start  
        #dR2dt *= torch.max(H1) 

        # scale for adaptation
        rsc1 = R*torch.max(Z).detach()
        rsc2 = R*torch.max(H1).detach()

        for _it,t in enumerate(t_):

            # dynamic input: multiply input with boxcar
            Z,_,_,_ = model.forw_conv(input_*boxcar[_it])
            
            # pre-activation
            dZdt[:,_it+1] = (Z + dh1dt[:,_it] - dR1dt[:,_it] - alpha_inh1[_it])/T
            
            # dynamics hidden layer1
            dh1dt[:,_it+1] = dh1dt[:,_it] + dt/tau_h * (-dh1dt[:,_it] + aet_net_2lay.sigmoid(dZdt[:,_it+1],model.sig_param[0]))      

            # adaptation term 1
            dR1dt[:,_it+1] = dR1dt[:,_it] + dt/tau_R * (-dR1dt[:,_it] + rsc1*dh1dt[:,_it+1])
            
            # pre-activation layer 2
            Z2 = (model.fc1(dh1dt[:,_it+1]) + dh2dt[:,_it+1] - dR2dt[:,_it]- alpha_inh2[_it])/T            
            # dynamics hidden layer1
            dh2dt[:,_it+1] = dh2dt[:,_it] + dt/tau_h * (-dh2dt[:,_it] + aet_net_2lay.sigmoid(Z2,model.sig_param[1]))      

            # # adaptation term 2
            dR2dt[:,_it+1] = dR2dt[:,_it] + dt/tau_R * (-dR2dt[:,_it] + rsc2*dh2dt[:,_it+1])          

            # output layer
            dOdt[:,_it+1] = model.actiout(model.fc2(dh2dt[:,_it+1]))

    return dZdt, dh1dt, dR1dt, dh2dt, dR2dt,dOdt
```

```python

plt.rcParams["figure.figsize"] = (6,3)

nn_dim_all = [[28,24,12,3],[28,48,12,3],[28,68,12,3]]

# nn_dim_ = [28,24,12,3]   # [quadrant size, number of hidden nodes, number of output nodes]
eta_ = 1e-2      # learning rate 0.02 -> smooth loss
mini_sz_ = 1        # mini batch size (1 = use SGD)
num_epo_ = 30

kl_reg_ = [0,1e-3,2]      # beta, p and sparsity layer that is penalized
      
sig_param = [[2, -2.5],[2,-2.5]] # sigmoid slope and shift in x direction
ortho_lambda = 0
ortho_lay = 2

# loss function & final layer activation (for binary crossentropy use sigmoid)
lossfun = [nn.MSELoss(), nn.Softmax(dim=-1)]

ortho_avg = np.zeros((len(nn_dim_all),2))
sparse_avg = np.zeros((len(nn_dim_all),2))

# store best model parameters
bm_params = np.zeros((50,5))

# worst model parameters
wm_params = np.zeros((50,5))



for j in np.arange(50):

    for i, nn_dim_ in enumerate(nn_dim_all):

        params = nn_dim_,eta_,mini_sz_,num_epo_,kl_reg_,sig_param,ortho_lambda, ortho_lay

        # initialize model and weights
        model_2layer = aet_net_2lay.net(params,lossfun)
        model_2layer = aet_net_2lay.init_params(model_2layer,weight_init='uni')
        optimizer = torch.optim.Adam(model_2layer.parameters(),lr=eta_)

        # train model

        model_2layer.to(DEVICE)
        loss_hist = model_2layer.train(optimizer,noise=False,print_loss=False)


        Z,H1,H2,O = model_2layer.forw_conv(data)

        ortho_2layer_lay1 = torch.mm(H1.T,H1)

        ortho_2layer_lay2 = torch.mm(H2.T,H2)
        ortho_avg[i,0] = ortho_2layer_lay1.mean()
        ortho_avg[i,1] = ortho_2layer_lay2.mean()

        sparse_avg[i,0] = ((torch.round(H1,decimals = 3) == 0).sum(1)/nn_dim_[1]).mean()
        sparse_avg[i,1] = ((torch.round(H2,decimals = 3) == 0).sum(1)/nn_dim_[2]).mean()
        # %matplotlib inline
        # plt.plot(np.arange(model_2layer.num_ep),loss_hist.detach().numpy())
        # plt.ylabel('loss')
        # plt.xlabel('epoch')

        # plt.show()
        # plt.savefig('loss.png')
        tau_h = 0.01
        tau_R = 0.1
        R = 3
        T = 0.05
        h_start = 0
        R_start = 0
        timevec = np.linspace(0,1,1000)
        alpha_params = [10,[2.5,0.5]]

        dyn_params = [tau_h,tau_R,R,T,h_start,R_start] 

        inp = data[[0]]*1.2 + data[[-1]]*.8

        Zt_, H1t_, R1t_, H2t_, R2t_, Ot_ = euler_dyn(model_2layer,inp, dyn_params,timevec,alpha_params,DEVICE,0)


        fig,ax = plt.subplots()
        ax.plot(Ot_.T.cpu().detach().numpy())
        ax.legend(('A(t)','E(t)','T(t)','Z(t)'),loc='lower right')
        ax.set_title('output')

#         fig,ax = plt.subplots(3,2)
#         ax[2,0].plot(Ot_.T.cpu().detach().numpy())
#         ax[2,0].legend(('A(t)','E(t)','T(t)','Z(t)'),loc='lower right')
#         ax[2,0].set_title('output')

#         # layer 2
#         ax[1,1].plot(R2t_.T.cpu().detach().numpy())
#         ax[1,1].set_title('Relaxation layer 2')
#         ax[1,0].plot(H2t_.T.cpu().detach().numpy())
#         ax[1,0].set_title('Hidden layer 2')

#         # layer 1
#         ax[0,1].plot(R1t_.T.cpu().detach().numpy())
#         ax[0,1].set_title('Relaxation layer 1')
#         ax[0,0].plot(H1t_.T.cpu().detach().numpy())
#         ax[0,0].set_title('Hidden layer 1')

#         if alpha_params[1][0]:
#             alpha_inh = 0.5*np.sin(2*np.pi*timevec*10)+0.5
#             ax[0,0].plot(alpha_inh,'k',linewidth=1,linestyle='-.')

#         if alpha_params[1][1]:
#             alpha_inh2 = 0.5*np.sin(2*np.pi*timevec*10-2*np.pi/10)+0.5

#             ax[2,0].plot(alpha_inh2,'k',linewidth=0.5,linestyle='-.')
#             ax[1,0].plot(alpha_inh2,'k',linewidth=0.5,linestyle='-.')



        %matplotlib inline
        fig.tight_layout()
        display(fig)
        # fig.savefig('dyn_nn_'+str(i)+'.png')
    
    x = input('best model?')
    
    if x != 'n':
        bm_params[j,:2] = ortho_avg[int(x),:] 
        bm_params[j,2:4] = sparse_avg[int(x),:] 
        bm_params[j,4] = int(x)


    x = input('worst model?')
    if x != 'n':
        wm_params[j,:2] = ortho_avg[int(x),:] 
        wm_params[j,2:4] = sparse_avg[int(x),:] 
        wm_params[j,4] = int(x)
        
    plt.close(fig)
```

```python
# Dynamics

# forward dynamics/discretization, euler integration
def new_euler_dyn(model,input_,params_,t_,alpha_params,DEVICE,inp_on):

    # inputs:
        # input_: input image
        # params_: hyperparameters
        # t_: time vector for integration
        # alpha_params: alpha frequency & amplitude

    # discretization & dynamics parameters
    tau_h,tau_R,R,T,h_start,R_start = params_


    # alpha frequency & amplitude
    _af,_aa = alpha_params

    with torch.no_grad():

        # initialize empty matrices
        dt = np.diff(t_)[0]
        dh1dt = (torch.zeros((model.dims[1],len(t_)+1))*h_start).to(DEVICE)
        dR1dt = (torch.zeros((model.dims[1],len(t_)+1))*R_start).to(DEVICE)
        dh2dt = (torch.zeros((model.dims[2],len(t_)+1))*h_start).to(DEVICE)
        dR2dt = (torch.zeros((model.dims[2],len(t_)+1))*R_start).to(DEVICE)
        dOdt = (torch.zeros((model.dims[-1],len(t_)+1))).to(DEVICE)
        dZdt = (torch.zeros((model.dims[1],len(t_)+1))).to(DEVICE)

        # alpha inhibition
        alpha_inh1 = _aa[0]*np.sin(2*np.pi*_af*t_)+_aa[0]
        alpha_inh2 = _aa[1]*np.sin(2*np.pi*_af*t_-2*np.pi/10)+_aa[1]

        # preactivation (dot product of input and first weight matrix)
        Z,H1,_,_ = model.forw_conv(input_)
        
        # create boxcar function to try different input onsets
        boxcar = np.zeros_like(t_)
        boxcar[inp_on:] = 1

        # adjust initial adaptation term (threshold)
        dR1dt[:,0] = R_start  
        dh1dt[:,0] = T
        #dR2dt *= torch.max(H1) 

        # scale for adaptation
        rsc1 = R#*torch.max(Z).detach()
        rsc2 = R*torch.max(H1).detach()

        for _it,t in enumerate(t_):

            # dynamic input: multiply input with boxcar
            Z,H,_,_ = model.forw_conv(input_*boxcar[_it])
            
            # pre-activation: logistic equation?
            dZdt[:,_it+1] = (Z + dh1dt[:,_it] - dR1dt[:,_it])/T
            
            # dynamics hidden layer1
            dh1dt[:,_it+1] = dh1dt[:,_it] + dt/tau_h * (-dh1dt[:,_it] + aet_net_2lay.sigmoid(dZdt[:,_it+1],model.sig_param[0]))      

            # adaptation term 1
            dR1dt[:,_it+1] = dR1dt[:,_it] + dt/tau_R * (-dR1dt[:,_it] +  rsc1 * dZdt[:,_it+1])
            
            # pre-activation layer 2
            Z2 = (model.fc1(dh1dt[:,_it+1]) + dh2dt[:,_it+1] - dR2dt[:,_it]- alpha_inh2[_it])/T            
            # dynamics hidden layer1
            dh2dt[:,_it+1] = dh2dt[:,_it] + dt/tau_h * (-dh2dt[:,_it] + aet_net_2lay.sigmoid(Z2,model.sig_param[1]))      

            # # adaptation term 2
            dR2dt[:,_it+1] = dR2dt[:,_it] + dt/tau_R * (-dR2dt[:,_it] + rsc2*dh2dt[:,_it+1])          

            # output layer
            dOdt[:,_it+1] = model.actiout(model.fc2(dh2dt[:,_it+1]))

    return dZdt, dh1dt, dR1dt, dh2dt, dR2dt,dOdt
```
